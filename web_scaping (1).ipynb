# Web Scrapping 

Importing libraries

```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

Using requests to fetch the html content of the website

```python
web_page=requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
soup=BeautifulSoup(web_page,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('h2'):
  print(i.text.strip())
```

```python
company_wrapper=soup.find_all('div',class_='company-content-wrapper')
```

Checking the number of content on a single page

```python
len(company_wrapper)
```

Trying if logic is working properly

```python
name=[]
rating=[]
reviews=[]
c_type=[]
h_q=[]
old=[]
employees=[]
tags=[]
for i in company_wrapper:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  reviews.append(i.find('a',class_='review-count').text.strip())
  c_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  h_q.append(i.find_all('p',class_='infoEntity')[1].text.strip())
  old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
  try:
    employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
  except:
    employees.append(np.nan)

  try:
    tags.append(i.find('a',class_='ab_chip').text.strip())
  except:
    tags.append(np.nan)
```

Creating a DataFrame using Pandas

```python
df={'name':name,'rating':rating,'Reviews':reviews,'Company_type':c_type,'Headquater':h_q,'Old':old,'#Employees':employees,'tags':tags}
```

```python
df=pd.DataFrame(df)
df
```

```python
name=[]
rating=[]
reviews=[]
c_type=[]
h_q=[]
old=[]
employees=[]
tags=[]
for j in range(1,333):
  url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
  web_page=requests.get(url).text
  soup=BeautifulSoup(web_page,'lxml')
  company_wrapper=soup.find_all('div',class_='company-content-wrapper')

  for i in company_wrapper:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    try:
      c_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      c_type.append(np.nan)
    try:
      h_q.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      h_q.append(np.nan)
    try:
      old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      old.append(np.nan)
    try:
      employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      employees.append(np.nan)
    try:
      tags.append(i.find('a',class_='ab_chip').text.strip())
    except:
      tags.append(np.nan)
  
df={'Name':name,'Rating':rating,'Reviews':reviews,'Company_type':c_type,'Headquarter':h_q,'Old':old,'No. of Employees':employees,'Tags':tags}
df=pd.DataFrame(df)
df
```

Exporting file as CSV

```python
df.to_csv(r'./list_of_companies.csv')
```
